```python
import requests
from io import StringIO
```

```python
url = 'https://github.com/rashakil-ds/5-Minutes-to-Pandas/raw/main/Datasets/bike.csv'
response = requests.get(url)
```

```python
csv_data = StringIO(response.text)
```
