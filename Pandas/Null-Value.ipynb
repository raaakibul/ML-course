```python
import pandas as pd 
import numpy as np 
```

```python
data = np.random.randn(10, 3)
for col in range(data.shape[1]):
    null_indices = np.random.choice(10,3, replace=False)
    data[null_indices, col] = np.nan
df = pd.DataFrame(data, columns=['A', 'B', 'C'])
```
