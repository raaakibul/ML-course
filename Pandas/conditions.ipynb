```python
import pandas as pd 
import requests
from io import StringIO
```

```python
url ='https://raw.githubusercontent.com/rashakil-ds/5-Minutes-to-Pandas/refs/heads/main/Datasets/Screen%20Time%20Data.csv'
response = requests.get(url)
```

```python
csv_data = StringIO(response.text)
```

```python
df = pd.read_csv(csv_data)
df
```

```python
# show the rows in the dataframe where week day is Wednesday and entertainment is 0
df[(df['Week Day']== 'Wednesday') & (df['Entertainment'] == 0)]
```

```python
# show the rows in the index name where Productivity is greater than 15

df['index'][df['Productivity'] > 15]
```

```python
df['Week Day'][df['Productivity'] > 15]
```

```python
# show the rows with the minimum productivity
df[df['Productivity'] == df['Productivity'].min()]
```

```python
# show the rows with the maximum Total screen time
df[df['Total Screen Time '] == df['Total Screen Time '].max()]
```

```python
# show the rows with the minimum Total screen time
df[df['Total Screen Time '] == df['Total Screen Time '].min()]
```

```python
df[['index', 'Week Day','Social Networking']] [df['Productivity']>15]
```
