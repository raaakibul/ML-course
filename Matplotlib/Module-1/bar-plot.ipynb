```python

# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 2, 5]

# Creating the bar plot
plt.bar(categories, values, color='purple')

# Adding title and labels
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')

# Displaying the plot
plt.show()
```

```python
# scatter plot 
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Creating the scatter plot
plt.scatter(x, y, color='blue')

# Adding title and labels
plt.title('Scatter Plot Example')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Displaying the plot
plt.show()
```

```python
import numpy as np
data = np.random.randn(1000)

plt.hist(data, bins=30, color='green')
plt.title('Histogram Example')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.show()
```

```python
# Data
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]

# Create a customized bar plot
plt.bar(categories, values, color='skyblue', width=0.6)

# Add gridlines
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Customized Bar Plot')

# Display the plot
plt.show()
```

```python
# Create a horizontal bar plot
plt.barh(categories, values, color='lightgreen')

# Add labels and title
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Plot')

# Display the plot
plt.show()
```

```python
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D']
group1 = [3, 7, 8, 5]
group2 = [4, 6, 9, 7]

# X-axis positions
x = np.arange(len(categories))

# Width of bars
width = 0.35

# Create grouped bar plot
plt.bar(x - width/2, group1, width, label='Group 1', color='lightblue')
plt.bar(x + width/2, group2, width, label='Group 2', color='orange')

# Add labels, legend, and title
plt.xticks(x, categories)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Grouped Bar Plot')
plt.legend()

# Display the plot
plt.show()
```

```python
# Create stacked bar plot
plt.bar(categories, group1, color='lightblue', label='Group 1')
plt.bar(categories, group2, bottom=group1, color='orange', label='Group 2')

# Add labels, legend, and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Plot')
plt.legend()

# Display the plot
plt.show()
```
